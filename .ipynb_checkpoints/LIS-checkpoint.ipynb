```python
def longest_increasing_subsequence(nums):
    if not nums:
        return []

    n = len(nums)
    dp = [1] * n          # length of LIS ending at i
    parent = [-1] * n     # to reconstruct the sequence

    max_len = 1
    max_idx = 0

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                parent[i] = j

        if dp[i] > max_len:
            max_len = dp[i]
            max_idx = i

    # Reconstruct sequence
    lis = []
    while max_idx != -1:
        lis.append(nums[max_idx])
        max_idx = parent[max_idx]

    return lis[::-1]


# Example
arr = [10, 9, 2, 5, 3, 7, 101, 18]
print(longest_increasing_subsequence(arr))
```
